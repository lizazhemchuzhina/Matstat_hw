```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
number_expirement = 100
legends = []
```

```python
def draw(windows, alphas):
    for i in windows:
        x = np.arange(i)
        for alpha in alphas:
            array_expect = [np.sort(np.random.exponential(alpha, i)) for _ in np.zeros(i)]
            y = sum(array_expect) / number_expirement    
            plt.plot(x,y)
            legends.append("alpha={}".format(alpha)) 
        plt.legend(legends)
        plt.xlabel("client position")
        plt.ylabel("time")
        plt.title("No more than two clients")
        plt.show()
        legends.clear()
```

```python
alphas = [0.1, 1, 2, 4, 10]
windows = [30, 100, 400, 1000]
```

```python
draw(windows, alphas)
```

```python
import heapq as hq
q_lengths = [60,200,800,2000]
```

```python
def hq_operations(number_window, alpha, queue_length):
    queue = (np.sort(np.random.exponential(alpha, number_window))).tolist()
    hq.heapify(queue)
    array_expect = [hq_add(queue,alpha) for _ in np.zeros(queue_length)]
    return np.array(array_expect)
```

```python
def hq_add(queue, alpha):
    top = hq.heappop(queue)
    hq.heappush(queue,np.random.exponential(alpha) + top)
    return top
```

```python
def draw_2(windows, alphas, queue_lengths):
    for i, j in zip(windows,queue_lengths):
        x = np.arange(j)
        for alpha in alphas:
            array_expect = [hq_operations(i, alpha, j) for _ in np.zeros(j)]
            y = sum(array_expect) / number_expirement    
            plt.plot(x,y)
            legends.append("alpha={}".format(alpha))   
        plt.legend(legends)
        plt.xlabel("client position")
        plt.ylabel("time")
        plt.title("A lot of clients with windows ={}".format(i))
        plt.show()
        legends.clear()
```

```python
draw_2(windows, alphas, q_lengths)
```
